```python
# Select all types
typelist=[]
for i in df[df['类型'].notnull()]['类型'].str.replace(' ','').str.split('/'):
    typelist.extend(i)
    
typelist=list(set(typelist))
print(typelist)
```

```python
datatype=df[df['类型'].notnull()][['电影名称','豆瓣评分','类型']]

type_lp=[]

def f1(data, typei):
    dic_type={}
    datai=data[data['类型'].str.contains(typei)]
    dic_type['typename']=typei
    dic_type['type_Ip_pre']=len(datai[datai['豆瓣评分']<4.3])/len(datai)
    dic_type['typecount']=len(datai)
    return (dic_type)
    
for i in typelist:
    d1=f1(datatype,i)
    type_lp.append(d1)

df_type_lp=pd.DataFrame(type_lp)
top20=df_type_lp.sort_values(by='type_Ip_pre',ascending=False).iloc[:20]
top20


```

```python
#Clean data
df_loc=df[df['制片国家/地区'].notnull()][['电影名称','豆瓣评分','制片国家/地区']]
df_loc=df_loc[df_loc['制片国家/地区'].str.contains('中国大陆')]


loc_lst=[]
for i in df_loc['制片国家/地区'].str.replace(' ','').str.split('/'):
    loc_lst.extend(i)

loc_lst= list(set(loc_lst))
loc_lst.remove('中国大陆')
loc_lst.remove('中国')
loc_lst.remove('台湾')
loc_lst.remove('香港')
print(loc_lst)
    
```

```python
df_loc_lst=[]

def f2(data,loci):
    country={}
    datai=data[data['制片国家/地区'].str.contains(loci)]
    country['loc']=loci
    country['loc_lp_pre']=len(datai[datai['豆瓣评分']<4.3])/len(datai)
    country['locount']=len(datai)
    return (country)
    
    

for i in loc_lst:
    d2=f2(df_loc,i)
    df_loc_lst.append(d2)
    
loctop=pd.DataFrame(df_loc_lst)
loctop=loctop[loctop['locount']>=3]
loctop20=loctop.sort_values(by='loc_lp_pre',ascending=False).iloc[:20]
loctop20
```

```python
df['主演人数']=df['主演'].str.split('/').str.len()

dfcount1=df[['主演人数','豆瓣评分']].groupby('主演人数').count()
dfcount2=df[['主演人数','豆瓣评分']][df['豆瓣评分']<4.3].groupby('主演人数').count()
dftable=pd.merge(dfcount1,dfcount2,left_index=True, right_index=True)
dftable.columns= ['电影数量','烂片数量']

dftable.reset_index(inplace=True)
dftable['主演人数分类']=pd.cut(dftable['主演人数'],[0,2,4,6,9,50],labels=['1-2人','3-4人','5-6人','7-9人','10人以上'])
dftable1=dftable[['主演人数分类','电影数量','烂片数量']].groupby('主演人数分类').sum()
dftable1['烂片比例']=dftable1['烂片数量']/dftable1['电影数量']
dftable1
```

```python
df2=df[(df['主演'].notnull())&(df['豆瓣评分']<4.3)]
actor=[]
for i in df2['主演'][df2['主演'].notnull()].str.replace(' ','').str.split('/'):
    actor.extend(i)
actor=list(set(actor))
print(len(actor))
```

```python
 film_actor=[]
df1=df[df['主演'].notnull()]

for i in actor:
    datai=df1[df1['主演'].str.contains(i)]
    
    if len(datai)>2:
        film={}
        film['主演名字']=i
        film['烂片数量']=len(datai[datai['豆瓣评分']<4.3])
        film['电影数量']=len(datai)
        film['烂片比例']=len(datai[datai['豆瓣评分']<4.3])/len(datai)
        film_actor.append(film)
        

film_bad_actor=pd.DataFrame(film_actor)
film_bad_actortop20=film_bad_actor.sort_values(by='烂片数量',ascending=False).iloc[:20]
film_bad_actortop20

   
```

```python
#Clean Data
df_year=df[['电影名称','导演','豆瓣评分','上映日期']][df['上映日期'].notnull()]
df_year['上映日期'] = df_year['上映日期'].str.replace(' ','').replace('暂无',None).replace('不详',None)
df_year['Year']=[x[:4] for x in df_year['上映日期']]
df_year=df_year[(df_year['导演'].notnull())&(df_year['Year'].notnull())]
df_year = df_year[df_year['Year'].str[0] == '2']  # 去除错误数据
df_year['Year'] = df_year['Year'].astype(np.int)   # 年份设置为整型
df_year.head()
```

```python
#Get Director list
director=[]
for i in df_year['导演'].str.split('/'):
    director.extend(i)
    
director=list(set(director))
print(director)
```

```python
film_dirctor=[]
for i in director:
    datai=df_year[df_year['导演'].str.contains(i)]
    if len(datai)>10:
        dir_list={}
        dir_list['dir']=i
        dir_list['dir_lp_pre']=len(datai[datai['豆瓣评分']<4.3])/len(datai)
        dir_list['dircount']=len(datai)
        film_dirctor.append(dir_list)

film_dir=pd.DataFrame(film_dirctor)
film_dir_top=film_dir.sort_values(by='dir_lp_pre',ascending=False)
film_dir_top
```
